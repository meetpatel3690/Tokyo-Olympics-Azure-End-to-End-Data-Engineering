```python
import pandas as pd
```

```python
# Conversion code for .xlsx to .csv for multiple files in a folder.

import os
from pathlib import Path

# Define input folder containing .xlsx files and output folder for .csv files
input_folder = r'C:\Users\meetp\OneDrive\Desktop\Projects\Project-6-DE-Azure\Olympic_Data_Excel'
output_folder = r'C:\Users\meetp\OneDrive\Desktop\Projects\Project-6-DE-Azure\Olympic_Data'

# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Get all .xlsx files in the input folder
xlsx_files = Path(input_folder).glob('*.xlsx')

# Convert each .xlsx file to .csv
for xlsx_file in xlsx_files:
    # Read the Excel file
    df = pd.read_excel(xlsx_file)
    
    # Create output filename with .csv extension
    csv_filename = xlsx_file.stem + '.csv'
    csv_path = os.path.join(output_folder, csv_filename)
    
    # Save as CSV
    df.to_csv(csv_path, index=False)
    print(f'Converted {xlsx_file.name} to {csv_filename}')

print('Conversion complete!')
```
